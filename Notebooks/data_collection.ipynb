```python
def print_nested_dict(data, indent=0):
    """
    Recursively prints a nested dictionary in a structured format.

    Args:
        data (dict): Dictionary to print.
        indent (int): Indentation level for formatting.
    """
    for key, value in data.items():
        if isinstance(value, dict):
            print(" " * indent + f"{key}:")
            print_nested_dict(value, indent + 4)
        else:
            print(" " * indent + f"{key}: {value}")

def test_full_stock_info(ticker: str):
    """
    Test function to fetch and print all stock information.

    Args:
        ticker (str): Stock symbol (e.g., 'AAPL')
    """
    fetcher = StockDataFetcher()
    
    result = fetcher.get_full_stock_info(ticker)

    print("\n" + "=" * 50)
    print(f"FULL STOCK DATA FOR: {ticker.upper()}")
    print("=" * 50 + "\n")

    # Check error first
    if result["error"]:
        print("Error:", result["error"])
        return

    # Print everything
    print_nested_dict(result)

# print(test_full_stock_info("AAPL"))
```
